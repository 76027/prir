```
import tensorflow as tf
from math import *
import numpy as np


class M_prostokatow:
    def __init__(self,a,b,n):
        self.a = a
        self.b = b
        self.n = n

    def f(self,x):
        return pow(1.3 * x + 2.1, 1/2) / log(x + 0.9)



    def oblicz(self):

        s = 0
        h = (self.b - self.a) / self.n
        for i in range(1,self.n):
            s += self.f(self.a) * h
            self.a += h
        return s



class M_trapezow:
    def __init__(self, a, b, n):
        self.a = a
        self.b = b
        self.n = n





    def f(self,x):
        return pow(1.3 * x + 2.1, 1/2) / log(x + 0.9)


    def oblicz(self):
        s = 0
        h = (self.b - self.a) / self.n
        for i in range(1,self.n+1):
            s += ((self.f(self.a) + self.f(self.a+h)) * h) / 2
            self.a += h
        return s

class M_Simpsona():
    def __init__(self, a, b, n):
        self.n = n
        self.a = a
        self.b = b



    def f(self, x):
        return pow(1.3*x+2.1,1/2)/log(x+0.9)

    def oblicz(self):
        s = 0
        st = 0
        h = (self.b - self.a) / self.n
        for i in range(1,self.n+1):
            x = self.a + i * h
            st += self.f(x - h / 2)
            if (i < self.n):
                s += self.f(x)

        s = h / 6 * (self.f(self.a) + self.f(self.b) + 2 * s + 4 * st)
        return s
```

Tworzę klasy z z zaimplemontowanymi metodami trapezów, prostokątów i Simpsona,obliczające pole pod całką (1.3 * x + 2.1)^1/2 / log(x + 0.9)

```
x = np.arange(1,10,1)
x = np.asarray(x)
print(x)
```

Tworzę macierz elementów z przedziału od 1 do 10 z krokiem 1

```
print("Metoda prostokatow")
met_pro = [M_prostokatow(i,i+1,3000).oblicz() for i in x]
met_pro = np.asarray(met_pro)
print(met_pro)

print("Metoda trapezow")
met_tra = [M_trapezow(i,i+1,3000).oblicz() for i in x]
met_tra = np.asarray(met_tra)
print(met_tra)

print("Metoda Simpsona")
met_sim = [M_Simpsona(i,i+1,3000).oblicz() for i in x]
met_sim = np.asarray(met_sim)
print(met_sim)
```

Wypisuje wyniki z działania funkcji w przedziałach [1:2] .... [9:10] przy 3000 iteracji


```
model_p = tf.keras.models.Sequential([tf.keras.layers.Dense(1, input_shape=(1, ))])

model_t = tf.keras.models.Sequential([tf.keras.layers.Dense(1, input_shape=(1, ))])

model_S = tf.keras.models.Sequential([tf.keras.layers.Dense(1, input_shape=(1, ))])

```

Tworzę modele dla każdej metody

```
#model_p.compile(loss='mse',optimizer=tf.optimizers.Adam(learning_rate=0.1))
model_p.compile(optimizer='sgd', loss='mean_squared_error')

#model_t.compile(loss='mse',optimizer=tf.optimizers.Adam(learning_rate=0.1))
model_t.compile(optimizer='sgd', loss='mean_squared_error')
#model_S.compile(loss='mse',optimizer=tf.optimizers.Adam(learning_rate=0.1))
model_S.compile(optimizer='sgd', loss='mean_squared_error')
```

Porównuję działanie dwóch optymizatorów, optymizator adam z rating rate =0.1, optymizator mean_squared_error wydaje się dokładniejszy w tym przykładzie.


```
model_p.fit(x,met_pro,epochs = 500)

model_t.fit(x,met_tra,epochs = 500)

model_S.fit(x,met_sim,epochs = 500)
```

Trenuję modele,po 500 epok

```
print("Przewidywania metody prostokatow")
p = model_p.predict(x)
print(p.flatten())
print("######Wynik z metody prostokatow#######")
print(met_pro)
print("###################")
srednia = 0
for i in range(len(x)):
  
  srednia += abs(p[i]-met_pro[i])
  print("Roznica miedzy wynikami: ",abs(p[i]-met_pro[i]))

srednia = srednia/len(x)
print("Srednia arytmetyczna bledow: ",srednia)
```

```
print("Przewidywania metody trapezow")
t = model_t.predict(x)
print(t.flatten())
print("######Wynik z metody trapezow#######")
print(met_tra)
print("###################")
srednia = 0
for i in range(len(x)):
  
  srednia += abs(t[i]-met_tra[i])
  print("Roznica miedzy wynikami: ",abs(p[i]-met_tra[i]))

srednia = srednia/len(x)
print("Srednia arytmetyczna bledow: ",srednia)
```

```
print("Przewidywania metody Simpsona")
s = model_S.predict(x)
print(s.flatten())
print("######Wynik z metody Simpsona#######")
print(met_sim)
print("###################")
srednia = 0
for i in range(len(x)):
  
  srednia += abs(t[i]-met_sim[i])
  print("Roznica miedzy wynikami: ",abs(p[i]-met_sim[i]))

srednia = srednia/len(x)
print("Srednia arytmetyczna bledow: ",srednia)
```
