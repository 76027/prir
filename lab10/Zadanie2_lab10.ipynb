```
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras import *
from math import *
 
```

```
def Funkcja_1d(x):
  return (x-2)*(x+1)*(x-4)

def Funkcja_2d(x,y):
  z = sin(sqrt((x*x)+(y*y)))/sqrt((x*x)+(y*y))
  return z
```

```
print("Zbiór x w funkcji 1d")
x1 = np.arange(-20,20,0.5,dtype=float)

print(x1)

print("Zbiór y w funkcji 1d")
y1 = []
for i in range(len(x1)):
    y1.append(float("{:.3f}". format(Funkcja_1d(x1[i]))))
y1 = np.array(y1)

#y1 = np.array([Funkcja_1d(i) for i in x1])

print(y1)

print("Zbiór x w funkcji 2d")

x2 = np.arange(-2, 2, 0.25, dtype=float)

print(x2)
print("Zbiór y w funkcji 2d")
y2 = np.arange(-2.25, 1.75, 0.25, dtype=float)

print(y2)

print("Zbiór z w funkcji 2d")
z =[]
for i in range(len(x2)):
    z.append(Funkcja_2d(x2[i],y2[i]))
z = np.array(z)
print(z)
```

```
# model1_d = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=(1,))
                               
# ])

model1_d = tf.keras.Sequential([
 keras.layers.Dense(units=1, input_shape=(1,)),
  keras.layers.Dense(units = 2, activation=keras.activations.sigmoid),
  keras.layers.Dense(units = 1)
])

model1_d.compile(loss='mse',optimizer=tf.optimizers.Adam(learning_rate=0.1))#(optimizer='sgd', loss='mean_squared_error')
```

```
model1_d.fit(x1, y1, epochs=1000)
```

```
print(model1_d.predict([4.0]))
```

Model dla funkcji 1d niezależnie od wytrenowania i liczebności zbioru treningowego nie jest zbyt dokładny

```
model_2d = keras.Sequential([
  keras.layers.Dense(units=2, input_shape=(2,)),
  keras.layers.Dense(units = 3, activation=keras.activations.sigmoid),
  keras.layers.Dense(units = 1)
])
model_2d.summary()
```

Tworzę model z jedną "ukrytą" warstwą

```

x_y = np.stack([x2,y2], axis=1)
model_2d.compile(
    optimizer = keras.optimizers.RMSprop(lr=0.02,rho=0.9,epsilon=None,decay=0),
    loss = 'mean_squared_error',
    metrics=['accuracy']
)
print(x_y[0])
```

```
model_2d.fit(x_y, z, epochs=100)
```
